```python
#Importando Biblioteca
import pandas as pd
```

```python
#Leitura de Arquivos
df1 = pd.read_excel("/content/drive/MyDrive/datasets/Aracaju.xlsx")
df2 = pd.read_excel("/content/drive/MyDrive/datasets/Fortaleza.xlsx")
df3 = pd.read_excel("/content/drive/MyDrive/datasets/Natal.xlsx")
df4 = pd.read_excel("/content/drive/MyDrive/datasets/Recife.xlsx")
df5 = pd.read_excel("/content/drive/MyDrive/datasets/Salvador.xlsx")
```

```python
# juntando todos os arquivos
df = pd.concat([df1,df2,df3,df4,df5])
```

```python
#Alterando o tipo de dado da coluna LojaID
df["LojaID"] = df["LojaID"].astype("object")
```

```python
#Consultar valores faltantes/nulos
df.isnull().sum()
```

```python
#Substituindo valores nulos pela media
df["Vendas"].fillna(df["Vendas"].mean(),inplace=True)
#Segunda opcao para trocar por zero
#df["Vendas"].fillna(0,inplace=True)
```

```python
df["Vendas"].mean()
```

```python
#Ja nao existia valores nulos
df.isnull().sum()
```

```python
df.sample(15)
```

```python
#apagando linhas valores nulos
df.dropna(inplace=True)
```

```python
#apagando linhas quando nulos em vendas
df.dropna(subset=["Vendas"], inplace=True)
```

```python
#apagando linhas que estejam todas as colunas vazias
df.dropna(how="all", inplace=True)
```

```python
#criando coluna de receita
df["Receita"]=df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
df["Receita/Vendas"]=df["Receita"]/df["Vendas"]
```

```python
#top 3 na coluna receita
df.nlargest(3,"Receita")
```

```python
#top worst 3 na coluna receita
df.nsmallest(3,"Receita")
```

```python
#Agrupamento por cidade
df.groupby("Cidade")["Receita"].sum()
```
